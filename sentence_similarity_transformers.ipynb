```python
from datasets import load_dataset
from sentence_transformers import SentenceTransformer, util, InputExample, losses
from sklearn.metrics import accuracy_score
from torch.utils.data import DataLoader
```

```python
dataset = load_dataset("paws", "labeled_final")
```

```python
train_data = dataset['train']
valid_data = dataset['validation']
test_data = dataset['test']
```

```python
train_data[:5]
```

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

def calculate_accuracy(data):
    embeddings1 = model.encode(data['sentence1'], convert_to_tensor=True)
    embeddings2 = model.encode(data['sentence2'], convert_to_tensor=True)

    similarities = util.pytorch_cos_sim(embeddings1, embeddings2).diag().cpu().tolist()
    predictions = [1 if sim >= 0.5 else 0 for sim in similarities]

    accuracy = accuracy_score(data['label'], predictions)
    return accuracy
```

```python
train_accuracy = calculate_accuracy(train_data)
print(f"Train Accuracy: {train_accuracy:.4f}")
```

```python
valid_accuracy = calculate_accuracy(valid_data)
print(f"Validation Accuracy: {valid_accuracy:.4f}")

test_accuracy = calculate_accuracy(test_data)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
train_examples = [InputExample(texts=[data['sentence1'], data['sentence2']], label=float(data['label'])) for data in train_data]
```

```python
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)
train_loss = losses.CosineSimilarityLoss(model)
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=3, warmup_steps=100)

model.save("fine_tuned_model")
fine_tuned_model = SentenceTransformer("fine_tuned_model")
```

```python
def calculate_accuracy(data):
    embeddings1 = fine_tuned_model.encode(data['sentence1'], convert_to_tensor=True)
    embeddings2 = fine_tuned_model.encode(data['sentence2'], convert_to_tensor=True)

    similarities = util.pytorch_cos_sim(embeddings1, embeddings2).diag().cpu().tolist()
    predictions = [1 if sim >= 0.5 else 0 for sim in similarities]

    accuracy = accuracy_score(data['label'], predictions)
    return accuracy

train_accuracy = calculate_accuracy(train_data)
print(f"Train Accuracy: {train_accuracy:.4f}")
```

```python
valid_accuracy = calculate_accuracy(valid_data)
print(f"Validation Accuracy: {valid_accuracy:.4f}")

test_accuracy = calculate_accuracy(test_data)
print(f"Test Accuracy: {test_accuracy:.4f}")
```
